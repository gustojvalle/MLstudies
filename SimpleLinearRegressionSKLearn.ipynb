```python
#Importing Libraries and setting the enviromnent

import pandas as pd 
import numpy as np 
import sklearn 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.linear_model import LinearRegression
```

```python
#Loading the dataset
boston = pd.read_csv('Boston.csv')
boston.head()

```

```python
boston.set_index('Unnamed: 0')
```

```python
#Analysing correlation of the features with seaborn
plt.figure(figsize=(16,10))
sns.heatmap(boston.corr(),annot=True)
plt.show
```

```python
#Defining the features to be used which will be 'lstat' and 'medv' very good correlation index.
x= boston['lstat'].values.reshape(-1,1)
y= boston['medv'].values
```

```python
#Defining the model to be used, in this case Linear due to the high correlation coeficient
model=LinearRegression()

#Fitting the model

model.fit(x,y)
```

```python
#Plotting to visualize the correlation
plt.figure(figsize=(12,16))
sns.regplot(x,y)
plt.xlabel('Amount of people living in the area with low income',fontsize=12)
plt.ylabel("MedVal in 10000's for occupied houses",fontsize=12)
plt.show()
```

```python
#Predicting random  values: 
pred=model.predict(np.array(range(40)).reshape(-1,1))
pred_x=np.array(range(40)).reshape(-1,1)
```

```python
#Plotting to visualize the correlation
plt.figure(figsize=(12,16))
sns.regplot(x,y)
sns.regplot(pred_x,pred)
plt.xlabel('Amount of people living in the area with low income',fontsize=12)
plt.ylabel("MedVal in 10000's for occupied houses",fontsize=12)
plt.show()
```
